# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

# Data Gathering

```python
df = pd.read_csv("./Breast Cancer Wisconsin (Diagnostic) Data Set.csv")
```

```python
df
```

```python
# id and unamed columns are not important so dropping it

df.drop(columns=["id","Unnamed: 32"],inplace=True)
```

# Understanding Dataset

```python
df.info()
```

```python
# Obtain descriptive statistics of DataFrame
df.describe()
```

```python
df.isna().sum()       # Checking for NULL values
```

```python
df.head()
```

```python
df["diagnosis"].unique()     # M --> Malignant(cancerous),  B --> Benign(non-cancerous)
```

```python
df["diagnosis"].value_counts()
```

```python
# Separating Dependent(x) and Independent(y) Variables.

x = df.drop(columns=["diagnosis"])     # Independent  
y = df[["diagnosis"]]                  # Dependent
```

```python
x.shape
```

```python
y.shape
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_train.head()
```

```python
x_test.head()
```

# Normalization and Standardization Technique

#### Normalization or Standardization is Important in KNN Algorithm


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

# KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

# Model Training

```python
knn.fit(x_train, y_train)
```

```python
ypred = knn.predict(x_test)
```

```python
ypred  # Predicted-Values
```

```python
y_test  # Actual-Values
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

conf_matrix = confusion_matrix(ypred, y_test)
print("Confusion Matrix:\n",conf_matrix)
```

```python
accuracy = accuracy_score(ypred, y_test)
print("Accuracy:", accuracy)
```

```python
report = classification_report(ypred, y_test)
print("Classification Report:\n\n",report)
```

# Hyperparameter Tunning

#### Hyperparameter-Tunning is Important in KNN Algorithm


### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

hyp = {'p':[1,2],    # p = parameter(distance metrics), 1 = Manhattan, 2 = Euclidean
          'n_neighbors':[2,3,4,5,6,7,8,9,10,25,30]}

knn_gscv = GridSearchCV(knn, hyp, cv=7)
```

```python
knn_gscv.fit(x_train, y_train)
```

```python
knn_gscv.best_params_  #shows th best choosen parameters from the given ones
```

```python
ypred_knn_gscv = knn_gscv.predict(x_test) 
ypred_knn_gscv
```

```python
acc_gscv = accuracy_score(ypred_knn_gscv, y_test)
print('Accuracy after GSCV Hyperparameter Tuning:', acc_gscv)
```

### RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

knn = KNeighborsClassifier()

hyp = {'p':[1,2],    # p = parameter(distance metrics), 1 = Manhattan, 2 = Euclidean
          'n_neighbors':[2,3,4,5,6,7,8,9,10,25,30]}

knn_rscv = RandomizedSearchCV(knn, hyp, cv=7)
```

```python
knn_rscv.fit(x_train, y_train)
```

```python
knn_rscv.best_params_  #shows th best choosen parameters from the given ones
```

```python
ypred_knn_rscv = knn_rscv.predict(x_test) 
ypred_knn_rscv
```

```python
acc_rscv = accuracy_score(ypred_knn_rscv, y_test)
print('Accuracy after RSCV Hyperparameter Tuning:', acc_rscv)
```

# Overall Accuracy

```python
print("Accuracy:", accuracy*100)
print('Accuracy after GSCV Hyperparameter Tuning :', acc_gscv*100)
print('Accuracy after RSCV Hyperparameter Tuning:', acc_rscv*100)
```
